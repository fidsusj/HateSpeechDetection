```python
import matplotlib.pyplot as plt
```

```python
# results: unbalanced, undersampled, oversampled
decision_tree = [92.58, 74.48, 79.46]
random_forrest = [93.20, 76.32, 86.98]
svm = [92.72, 77.39, 78.59]
logistic_regression = [93.05, 77.10, 78.19]
lstm = [93.90, 93.90]

x = [1, 2, 3]
labels = ["Unbalanced", "Undersampled", "Oversampled"]

plt.figure(figsize=(12, 12))

size=75
marker="o"
plt.scatter(x, decision_tree, marker=marker, s=size, label="Decision Tree")
plt.scatter(x, random_forrest, marker=marker, s=size, label="Random Forrest")
plt.scatter(x, svm, marker=marker, s=size, label="SVM")
plt.scatter(x, logistic_regression, marker=marker, s=size, label="Logistic Regression")
plt.scatter([1, 2], lstm, marker=marker, s=size, label="LSTM")

plt.xticks(x, labels)
plt.ylim(50, 100)
plt.legend()
plt.show()
```
