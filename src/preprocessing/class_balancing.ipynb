```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from matplotlib import pyplot as plt
from collections import Counter
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
```

```python
counter = Counter(y)
print(counter)
```

```python
for label, _ in counter.items():
    row_ix = np.where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
# transform dataset using SMOTE
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
smote_X, smote_y = oversample.fit_resample(X, y)
```

```python
smote_counter = Counter(smote_y)
print(smote_counter)

for label, _ in smote_counter.items():
    row_ix = np.where(smote_y == label)[0]
    plt.scatter(smote_X[row_ix, 0], smote_X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
# instead of using SMOTE to do oversampling, we can just undersample the minority class
from imblearn.under_sampling import RandomUnderSampler

undersample = RandomUnderSampler()
under_X, under_y = undersample.fit_resample(X, y)
print(Counter(under_y))
```
